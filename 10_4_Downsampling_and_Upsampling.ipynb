```python
def subsample(x_in):
    x_out = x_in[::2, ::2]
    return x_out
```

```python
# Now let's try max-pooling
def maxpool(x_in):
  # TO DO -- write the maxpool routine
    x_out = np.zeros((x_in.shape[0]//2, x_in.shape[1]//2), dtype=x_in.dtype)
    for i in range(x_in.shape[0]//2):
        for j in range(x_in.shape[1]//2):
            block = x_in[i*2:(i+1)*2, j*2:(j+1)*2]
            if block.size > 0:
                x_out[i, j] = np.max(block)
    return x_out
```

```python
# Finally, let's try mean pooling
def meanpool(x_in):
  # TO DO -- write the meanpool routine
    x_out = np.zeros((x_in.shape[0]//2, x_in.shape[1]//2), dtype=x_in.dtype)
    for i in range(x_in.shape[0]//2):
        for j in range(x_in.shape[1]//2):
            block = x_in[i*2:(i+1)*2, j*2:(j+1)*2]
            if block.size > 0:
                x_out[i, j] = np.mean(block)
    return x_out
```

```python
# Let's first use the duplication method
def duplicate(x_in):
    # TO DO -- write the duplication routine
    x_out = np.zeros((x_in.shape[0] * 2, x_in.shape[1] * 2), dtype=x_in.dtype)
    for i in range(x_out.shape[0]):
        for j in range(x_out.shape[1]):
            x_out[i, j] = x_in[i//2, j//2]
    return x_out
```

```python
# Now let's try max pooling back up
# The input x_high_res is the original high res image, from which you can deduce the position of the maximum index
def max_unpool(x_in, x_high_res):
    x_out = np.zeros(( x_in.shape[0]*2, x_in.shape[1]*2 ))
    for i in range(x_in.shape[0]):
        for j in range(x_in.shape[1]):
            block = x_high_res[2*i:2*(i+1), 2*j:2*(j+1)]
            max_index = np.unravel_index(np.argmax(block, axis=None), block.shape) # Find the index of the max value within this block
            x_out[2*i + max_index[0], 2*j + max_index[1]] = x_in[i, j] # Place the value from x_in into the position of the max value in the block within x_ou
    return x_out
```

Finally, we'll try upsampling using bilinear interpolation.  We'll treat the positions off the image as zeros by padding the original image and round fractional values upwards using np.ceil()

```python
def bilinear(x_in):
    x_out = np.zeros(( x_in.shape[0]*2, x_in.shape[1]*2 ))
    x_in_pad = np.zeros((x_in.shape[0]+1, x_in.shape[1]+1))
    x_in_pad[0:x_in.shape[0],0:x_in.shape[1]] = x_in
    # TO DO -- write the duplication routine

    for i in range(x_out.shape[0]):
        for j in range(x_out.shape[1]):
            # Compute the indices of the top-left corner of the cell in the input matrix that corresponds to the current pixel in the output matrix
            x = i // 2
            y = j // 2
            
            # Calculate the weights for the interpolation
            a = i / 2 - x
            b = j / 2 - y
            
            # Perform the bilinear interpolation using the four surrounding pixels
            value = (1 - a) * (1 - b) * x_in_pad[x, y] + \
                    a * (1 - b) * x_in_pad[x + 1, y] + \
                    (1 - a) * b * x_in_pad[x, y + 1] + \
                    a * b * x_in_pad[x + 1, y + 1]
            
            # Assign the interpolated value to the current pixel, rounding upwards
            x_out[i, j] = np.ceil(value)
    return x_out
```
